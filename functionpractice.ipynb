```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure
```

```python
#Objective Function
def u(c1, c2, c3, c4):
    return math.log(c1) + ß*math.log(c2) + (ß**2)*math.log(c3) + (ß**3)*math.log(c4)
#Constraint
def c(c1, c2, c3, c4):
    return c1 + c2/π + c3/(π**2) + c4/(π**3)
```

```python
#Parameters:
ß = 1.00
π = 1.05
c0 = 100
t = 100
```

```python
#Relevant Functions:
def consumption_function(i):
    return ((ß**(i-1)*π**(i-1))/sum(ß**(x-1) for x in range(1, t+1))*c0)

def capital_function(i):
    if i == 1:
        return c0
    else:
        return π*(capital_function(i-1)-consumption_function(i-1))
```

```python
 time = np.linspace(1, t, t)

#Computing the path of Friday's consumption across time:
 time_consumption = np.zeros(len(time))
 time_consumption = [consumption_function(i) for i in time]


#Computing the Crusonia plant's mass across time:
 time_capital = np.zeros(len(time))
 time_capital = [capital_function(i) for i in time]
 
 print(time)
 print (time_consumption)
 print (time_capital)

summarydf = pd.DataFrame(data=np.column_stack((time,time_consumption, time_capital)),columns=['Period','Consumption', 'Capital'])
print(summarydf)
```
