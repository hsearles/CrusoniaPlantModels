```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure
```

```python
#Objective Function
def u(c1, c2, c3, c4):
    return math.log(c1) + ß*math.log(c2) + (ß**2)*math.log(c3) + (ß**3)*math.log(c4)
#Constraint
def c(c1, c2, c3, c4):
    return c1 + c2/π + c3/(π**2) + c4/(π**3)
```

```python
#Parameters:
ß = 1/1.05
π = 1.05
c0 = 4
t = 100
```

```python
#Relevant Functions:
def consumption_function(i):
    return ((ß**(i-1)*π**(i-1))/sum(ß**(x-1) for x in range(1, t+1))*c0)

def consumption_function2(i):
    return (c0/sum(ß**(i-1) for i in range(1, t+1)))*(ß*π)**(i-1)


def capital_function(i):
    if i == 1:
        return c0
    else:
        return π*(capital_function(i-1)-consumption_function(i-1))

def income_function(i):
    return capital_function(i) / sum(π**(1-i) for i in range(1, (t-i)+2))
```

```python
 time = np.linspace(1, t, t)
 time2 = [i for i in range(1,t+1)]


#Computing the path of Friday's consumption across time:
 time_consumption = np.zeros(len(time))
 time_consumption = [consumption_function(i) for i in time]


#Computing the Crusonia plant's mass across time:
 time_capital = np.zeros(len(time))
 time_capital = [capital_function(i) for i in time]
 
time_income = [income_function(i) for i in time2]

 print(time)
 print (time_consumption)
 print (time_capital)

summarydf = pd.DataFrame(data=np.column_stack((time,time_consumption, time_capital, time_income)),columns=['Period','Consumption', 'Capital', 'Income'])
print(summarydf)
```

```python
plt.plot(time, time_consumption, color = 'deepskyblue')

font1 = {'family':'MS Reference Sans Serif', 'color':'darkslategray','size':14}
font2 = {'family':'MS Reference Specialty', 'color':'black','size':17}
plt.ylim(ymin=0) 

plt.xlabel("Time", fontdict = font1)
plt.ylabel("Consumption", fontdict = font1)
plt.title("Friday's Consumption Over Time", fontdict = font2)

plt.show()
```
